```python
# ==========================================
# Stress‚ÄìStrain Data Analysis (Colab Script)
# ==========================================

# Step 0 ‚Äî Import required libraries
import pandas as pd

# ------------------------------------------
# Step 1 ‚Äî Load CSV file and display first 5 rows
# ------------------------------------------

# If file is already uploaded to Colab, place it in the same directory
file_name = "stress_strain_practice.csv"

df = pd.read_csv(file_name)

print("First 5 rows of dataset:")
print(df.head())


# ------------------------------------------
# Step 2 ‚Äî Calculate Stress (MPa)
# Stress = Force / Area
# Units: N/mm¬≤ = MPa (already correct)
# ------------------------------------------

df["Stress_MPa"] = df["Force_N"] / df["Cross_Section_Area_mm2"]


# ------------------------------------------
# Step 3 ‚Äî Calculate Strain (%)
# Strain (%) = (Elongation / Original Length) * 100
# ------------------------------------------

df["Strain_percent"] = (
    df["Elongation_mm"] / df["Original_Length_mm"]
) * 100


print("\nData with Stress and Strain columns:")
print(df.head())


# ------------------------------------------
# Step 4 ‚Äî Group by Material and compute averages
# ------------------------------------------

material_avg = (
    df.groupby("Material")[["Stress_MPa", "Strain_percent"]]
    .mean()
    .reset_index()
)

print("\nAverage Stress and Strain by Material:")
print(material_avg)


# ------------------------------------------
# Step 5 ‚Äî Identify strongest material
# (highest average stress)
# ------------------------------------------

strongest_material = material_avg.loc[
    material_avg["Stress_MPa"].idxmax(), "Material"
]

print(f"\nStrongest material (highest average stress): {strongest_material}")


# ------------------------------------------
# Step 6 ‚Äî Summary statistics
# min, max, mean stress per material
# ------------------------------------------

stress_summary = (
    df.groupby("Material")["Stress_MPa"]
    .agg(["min", "max", "mean"])
    .reset_index()
)

print("\nStress Summary (min, max, avg):")
print(stress_summary)


# ------------------------------------------
# Step 7 ‚Äî Export results to new CSV files
# ------------------------------------------

# Save full dataset with new columns
df.to_csv("stress_strain_with_calculations.csv", index=False)

# Save grouped averages
material_avg.to_csv("material_averages.csv", index=False)

# Save summary stats
stress_summary.to_csv("stress_summary.csv", index=False)

print("\nFiles exported:")
print("- stress_strain_with_calculations.csv")
print("- material_averages.csv")
print("- stress_summary.csv")
```

```python
# ==========================================
# Stress‚ÄìStrain Analysis with Visualization
# ==========================================

# Step 0 ‚Äî Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------------------
# Step 1 ‚Äî Load CSV with error handling
# ------------------------------------------

file_name = "stress_strain_practice.csv"

try:
    df = pd.read_csv(file_name)
    print("File loaded successfully!\n")
except FileNotFoundError:
    print("ERROR: File not found. Upload it first using files.upload()")
    raise

print("First 5 rows:")
print(df.head())


# ------------------------------------------
# Step 2 ‚Äî Check and handle missing values
# ------------------------------------------

print("\nMissing values per column:")
print(df.isnull().sum())

# Drop rows with missing critical numeric data
critical_cols = [
    "Force_N",
    "Cross_Section_Area_mm2",
    "Elongation_mm",
    "Original_Length_mm",
    "Material"
]

df = df.dropna(subset=critical_cols)

print("\nAfter removing missing values, remaining rows:", len(df))


# ------------------------------------------
# Step 3 ‚Äî Calculate Stress (MPa)
# ------------------------------------------

df["Stress_MPa"] = df["Force_N"] / df["Cross_Section_Area_mm2"]


# ------------------------------------------
# Step 4 ‚Äî Calculate Strain (%)
# ------------------------------------------

df["Strain_percent"] = (
    df["Elongation_mm"] / df["Original_Length_mm"]
) * 100


# ------------------------------------------
# Step 5 ‚Äî Highlight tests where stress > 200 MPa
# ------------------------------------------

high_stress_tests = df[df["Stress_MPa"] > 200]

print("\n‚ö† Tests with Stress > 200 MPa:")
print(high_stress_tests)

# Save separately
high_stress_tests.to_csv("high_stress_tests.csv", index=False)


# ------------------------------------------
# Step 6 ‚Äî Group by Material (average values)
# ------------------------------------------

material_avg = (
    df.groupby("Material")[["Stress_MPa", "Strain_percent"]]
    .mean()
    .reset_index()
)

print("\nAverage Stress & Strain by Material:")
print(material_avg)


# ------------------------------------------
# Step 7 ‚Äî Strongest material
# ------------------------------------------

strongest_material = material_avg.loc[
    material_avg["Stress_MPa"].idxmax(), "Material"
]

print(f"\nüèÜ Strongest Material: {strongest_material}")


# ------------------------------------------
# Step 8 ‚Äî Stress summary statistics
# ------------------------------------------

stress_summary = (
    df.groupby("Material")["Stress_MPa"]
    .agg(["min", "max", "mean"])
    .reset_index()
)

print("\nStress Summary:")
print(stress_summary)


# ------------------------------------------
# Step 9 ‚Äî Visualization (Bar Chart)
# Average Stress vs Material
# ------------------------------------------

plt.figure(figsize=(8,5))

plt.bar(material_avg["Material"], material_avg["Stress_MPa"])

plt.xlabel("Material Type")
plt.ylabel("Average Stress (MPa)")
plt.title("Average Stress by Material")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()


# ------------------------------------------
# Step 10 ‚Äî Export results
# ------------------------------------------

df.to_csv("stress_strain_with_calculations.csv", index=False)
material_avg.to_csv("material_averages.csv", index=False)
stress_summary.to_csv("stress_summary.csv", index=False)

print("\nFiles exported:")
print("- stress_strain_with_calculations.csv")
print("- material_averages.csv")
print("- stress_summary.csv")
print("- high_stress_tests.csv")
```
